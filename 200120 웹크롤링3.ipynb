```python
import requests
import lxml.html
import re
import time # 시간지연기능 추가
import sqlite3
con=sqlite3.connect('c:/sqlite/testDB')
cur=con.cursor()
cur.execute('drop table if exists ebook;') #테이블 있으면 지운다.
cur.execute('create table ebook(url text,title text, price text, contents text)')
session=requests.Session()
res=session.get('http://www.hanbit.co.kr/store/books/new_book_list.html')
root=lxml.html.fromstring(res.content)
root.make_links_absolute(res.url)
for a in root.cssselect('.book_tit a'):
    url=a.get('href')
    print('URL :',url)
    #time.sleep(1)
    res=session.get(url)
    root=lxml.html.fromstring(res.content)
    title=root.cssselect('.store_product_info_box h3')[0].text
    print('제목 :',title)
    price=root.cssselect('.pbr strong')[0].text
    print('가격 :',price)
    print('목차 :')
    contents=''
    for p in root.cssselect('#tabs_3 .detail_conbox p'):
        content=p.text
        if content!=None:
            content=re.sub('\s+','_',content)
        if content==None or content=='':
            continue
        contents+=content+'\n'
    cur.execute('insert into ebook values(:url,:title,:price,:contents)',{'url':url,'title':title,'price':price,'contents':contents})
    print(contents)
    print('==========================================\n')
con.commit()
con.close()
```
